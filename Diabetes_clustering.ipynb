# Read File


The dataset represents 10 years (1999-2008) of clinical care at 130 US hospitals and integrated delivery networks. It includes over 50 features representing patient and hospital outcomes. Information was extracted from the database for encounters that satisfied the following criteria.
1. It is an inpatient encounter (a hospital admission).
2. It is a diabetic encounter, that is, one during which any kind of diabetes was entered to the system as a diagnosis.
3. The length of stay was at least 1 day and at most 14 days.
4. Laboratory tests were performed during the encounter.
5. Medications were administered during the encounter.

The data contains such attributes as patient number, race, gender, age, admission type, time in hospital, medical specialty of admitting physician, number of lab test performed, HbA1c test result, diagnosis, number of medication, diabetic medications, number of outpatient, inpatient, and emergency visits in the year before the hospitalization, etc.

# Model Building

First, I am going to create dummy variables for features that have text.

```python
diabete= pd.get_dummies(diabete,drop_first=True)
```

Now I am going to import my kmeans cluster

```python
diabete.columns
```

```python
diabete.head()
```

```python
from sklearn.cluster import KMeans
```

I amn going to create two clusters: Readmitted or not

```python
kmean= KMeans(n_clusters=2)
```

```python
kmean.fit(diabete.drop('readmitted_YES',axis=1))
```

Lets look at its cluster center

```python
kcluster_center= kmean.cluster_centers_
```

```python
kcluster_center
```

Lets see what was the prediction it came up with

```python
km_pred = kmean.fit_predict(diabete.drop('readmitted_YES',axis=1))
```

```python
km_pred
```

Putting the values in a dataframe

```python
diabete['Cluster'] = km_pred
```

```python
diabete.head()
```

Lets Evaluate our test. Noramlly, for clustering, we cannot evaluate our prediction because we do not have a depedent variable but since we want to cluster this into two groups, it will give us an idea on how our model did

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(diabete['readmitted_YES'],km_pred))
print(confusion_matrix(diabete['readmitted_YES'],km_pred))
```
